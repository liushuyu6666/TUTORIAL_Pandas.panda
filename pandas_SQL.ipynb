```select * from table where condition1 ```

````python
data[data.security_id == 1996480]
````

## group by
count

````python
data.groupby(['security_id']).size().to_frame('count').reset_index()
````

## having
```select security_id, count(*) as cnt from table group by security_id having cnt > 20```

````python
data.groupby('security_id').filter(lambda x : len(x) > 10).groupby('security_id').size()
````

## in

````python
data[data.security_id.isin([1954065, 1954165])].head(5)
````

## not in

````python
data[~data.security_id.isin([1954065, 1954165])].head(5)
````

## union
```select * from table where date < '2000-01-01' union all select * from talbe where date > '2010-01-01'```

````python
pd.concat([data[data.date < '2000-01-01'], data[data.date > '2010-01-01']])
````

## join
```select * from table join dim on table.security_id = dim.company_id where table.security_id = 1954065```

````python
data.merge(dim[data.security_id == 1954065], left_on='security_id', right_on='company_id', how='inner')
````
